```python
import numpy as np
import matplotlib.pyplot as plt
import numdifftools as nd
import qmcpy as qp
```

```python
d=2
n=2 ** 8

uniform = qp.IIDStdUniform(dimension=d, seed=0)
samples0 = uniform.gen_samples(n)

halton = qp.Halton(d, randomize=False)
samples1 = halton.gen_samples(n)

sobol = qp.Sobol(d, randomize=False)
samples2 = sobol.gen_samples(n)

lattice = qp.Lattice(d, randomize=False)
samples3 = lattice.gen_samples(n)

fig, ax = plt.subplots(2, 2, figsize=[10, 10])
ax[0, 0].plot(samples0[:, 0], samples0[:, 1], '.', label='Random points')
ax[0, 1].plot(samples1[:, 0], samples1[:, 1], '.', label='Halton points')
ax[1, 0].plot(samples2[:, 0], samples2[:, 1], '.', label='Sobol points')
ax[1, 1].plot(samples3[:, 0], samples3[:, 1], '.', label='Lattice points')
for a in ax.flatten():
    a.legend(loc=2)
    a.axis('scaled')

plt.show()
```
